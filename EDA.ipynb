```python
# import necessary libraries for EDA
import pandas as pd


# read in the data
data_movie_reviews = pd.read_csv('/workspaces/IDS706_Final_Project_klap/data/combined_movie_reviews.csv')
```

```python


print(data_movie_reviews.head(10)) # reading top 10 samples of the data
```

```python
print(f"there are {data_movie_reviews.shape[0]} rows and {data_movie_reviews.shape[1]} columns in the data")
```

```python
# performing EDA        
print(data_movie_reviews.info())
```

```python
# what are the values of label
print(data_movie_reviews['label'].value_counts()) 

# plot a pie chart to show positive and negative reviews
data_movie_reviews['label'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(6, 6))

print(f"the data is balanced as there are {data_movie_reviews['label'].value_counts()[0]} positive reviews and {data_movie_reviews['label'].value_counts()[1]} negative reviews")
```

```python
# plot a graph of the count of imdb and rotten tomatoes reviews
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
sns.countplot(x='source', data=data_movie_reviews)
plt.title('Count of IMDB and Rotten Tomatoes Reviews')
plt.show()

print(f"The count of IMDB and rotten tomatoes revies are {data_movie_reviews['source'].value_counts()[0]} and {data_movie_reviews['source'].value_counts()[1]} respectively")



```
